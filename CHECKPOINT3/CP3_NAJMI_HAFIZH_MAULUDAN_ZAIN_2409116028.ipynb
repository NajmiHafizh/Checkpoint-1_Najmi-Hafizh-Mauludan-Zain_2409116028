```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("slepp fixx.csv")
df
```

```python
print(df.columns)
```

```python
# Scatter Plot (Seaborn)
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Stress Level', y='Sleep Duration',  data=df)
plt.title('Korelasi Stres dan Skor Kualitas Tidur')
plt.xlabel('Stress Level')
plt.ylabel('Sleep Duration')
plt.show()
```

```python
# Histogram (Seaborn)
plt.figure(figsize=(8, 5))
sns.histplot(df['Sleep Duration'], bins=20, kde=True, color='blue')
plt.title('Distribusi Durasi Tidur')
plt.xlabel('Sleep Duration')
plt.ylabel('Frekuensi')
plt.show()
```

```python
# Boxplot (Seaborn)
plt.figure(figsize=(8, 5))
sns.boxplot(x=pd.qcut(df['Daily Steps'], q=3, labels=["Rendah", "Sedang", "Tinggi"]), y=df['Sleep Duration'], palette='pastel')
plt.title('Pengaruh Aktivitas Fisik terhadap Kualitas Tidur')
plt.xlabel('Daily Steps')
plt.ylabel('Sleep Duration')
plt.xticks(rotation=0)
plt.show()
```

```python
# Line Chart (Seaborn)
df['Age'] = df['Gender_and_Age'].str.extract('(\d+)').astype(int)
df['Gender'] = df['Gender_and_Age'].str.extract('([A-Za-z]+)')

plt.figure(figsize=(10, 5))
sns.lineplot(x='Age', y='Daily Steps', hue='Gender', data=df, marker='o', ci=None)
plt.title('Pola Jumlah Langkah Harian Berdasarkan Usia dan Gender')
plt.xlabel('Usia')
plt.ylabel('Daily Steps')
plt.grid()
plt.show()
```

```python
# 5. Heatmap: Korelasi antara Faktor Gaya Hidup dan Skor Kualitas Tidur
plt.figure(figsize=(10,8))
corr = df[['Sleep Duration', 'Stress Level', 'Daily Steps', 'Heart Rate']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Korelasi Faktor Gaya Hidup dan Durasi Tidur")
plt.show()
```

Berdasarkan hasil visualisasi, berikut adalah beberapa insight yang dapat diambil serta tindakan yang dapat dilakukan:

1.Korelasi Tingkat Stres dan Durasi Tidur (Scatter Plot)

- Ada pola negatif: semakin tinggi tingkat stres, semakin rendah durasi tidur.
- Tindakan: Rekomendasikan manajemen stres, seperti meditasi, olahraga ringan, atau jadwal tidur yang lebih teratur.

2.Distribusi Durasi Tidur (Histogram)

- Mayoritas individu tidur di rentang 5-7 jam, yang sedikit di bawah rekomendasi (7-9 jam).
- Tindakan: Perlu edukasi tentang pentingnya tidur cukup, terutama bagi kelompok yang tidur <6 jam.

3.Pengaruh Aktivitas Fisik terhadap Tidur (Boxplot)

- Kelompok dengan aktivitas fisik tinggi cenderung memiliki durasi tidur lebih lama dibandingkan yang rendah.
- Tindakan: Dorong peningkatan aktivitas fisik, seperti berjalan 8.000+ langkah sehari, untuk meningkatkan kualitas tidur.

4.Jumlah Langkah Harian vs. Durasi Tidur (Line Plot)

- Tidak ada hubungan linier yang kuat, tetapi individu dengan langkah lebih banyak cenderung memiliki durasi tidur yang stabil.
- Tindakan: Fokus pada keseimbangan antara aktivitas fisik dan waktu tidur.

5.Korelasi Faktor Gaya Hidup terhadap Tidur (Heatmap)

- Tingkat stres memiliki korelasi negatif yang kuat dengan durasi tidur.
Jumlah langkah dan detak jantung memiliki korelasi lemah, menandakan faktor ini mungkin kurang berpengaruh dibanding stres.
- Tindakan: Prioritaskan strategi pengurangan stres dibandingkan hanya meningkatkan aktivitas fisik.
